```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Données
data = np.array([
    [5, 4],   # ind1
    [4, 5],   # ind2
    [1, -2],  # ind3
    [0, -3]   # ind4
])

labels = ['ind1', 'ind2', 'ind3', 'ind4']

print("=== EXERCICE 1 : Classification K-moyennes ===\n")
print("Données initiales :")
for i, (point, label) in enumerate(zip(data, labels)):
    print(f"{label}: ({point[0]}, {point[1]})")

# Centres initiaux : ind1 et ind2
initial_centers = np.array([
    [5, 4],   # ind1
    [4, 5]    # ind2
])

print(f"\nCentres initiaux :")
print(f"Centre 1: ({initial_centers[0][0]}, {initial_centers[0][1]})")
print(f"Centre 2: ({initial_centers[1][0]}, {initial_centers[1][1]})")

# Application de K-moyennes avec centres initiaux
kmeans = KMeans(n_clusters=2, init=initial_centers, n_init=1, random_state=42, max_iter=300)
cluster_labels = kmeans.fit_predict(data)

print(f"\n=== RÉSULTATS ===")
print(f"Partition finale :")
for i, (label, cluster) in enumerate(zip(labels, cluster_labels)):
    print(f"{label}: Classe {cluster + 1}")

# Centres finaux
final_centers = kmeans.cluster_centers_
print(f"\nCentres finaux :")
for i, center in enumerate(final_centers):
    print(f"Centre {i+1}: ({center[0]:.1f}, {center[1]:.1f})")

# Calcul de l'inertie intra-classe
inertia = kmeans.inertia_
print(f"\nInertie intra-classe: {inertia:.1f}")

# Vérification manuelle du calcul d'inertie
print(f"\n=== VÉRIFICATION MANUELLE ===")
manual_inertia = 0
for i in range(2):  # Pour chaque classe
    class_points = data[cluster_labels == i]
    center = final_centers[i]
    class_inertia = np.sum((class_points - center) ** 2)
    print(f"Inertie classe {i+1}: {class_inertia:.1f}")
    manual_inertia += class_inertia

print(f"Inertie totale (calcul manuel): {manual_inertia:.1f}")

# 3) Calcul du pourcentage d'inertie expliquée
# Calcul de l'inertie totale du nuage de points
centroid = np.mean(data, axis=0)
total_inertia = np.sum((data - centroid) ** 2)

print(f"\n=== QUESTION 3 ===")
print(f"Centre de gravité du nuage: ({centroid[0]:.1f}, {centroid[1]:.1f})")
print(f"Inertie totale du nuage: {total_inertia:.1f}")

# L'inertie expliquée = inertie totale - inertie intra-classe
explained_inertia = total_inertia - inertia
percentage_explained = (explained_inertia / total_inertia) * 100

print(f"Inertie expliquée par la partition: {explained_inertia:.1f}")
print(f"Pourcentage d'inertie expliquée: {percentage_explained:.1f}%")

# Visualisation
plt.figure(figsize=(10, 8))

# Points colorés selon leur classe
colors = ['red', 'blue']
for i in range(2):
    class_points = data[cluster_labels == i]
    plt.scatter(class_points[:, 0], class_points[:, 1], 
               c=colors[i], s=100, alpha=0.7, label=f'Classe {i+1}')

# Centres finaux
plt.scatter(final_centers[:, 0], final_centers[:, 1], 
           c='black', marker='x', s=200, linewidths=3, label='Centres')

# Centres initiaux
plt.scatter(initial_centers[:, 0], initial_centers[:, 1], 
           c='green', marker='s', s=100, alpha=0.7, label='Centres initiaux')

# Étiquettes des points
for i, (point, label) in enumerate(zip(data, labels)):
    plt.annotate(label, (point[0], point[1]), xytext=(5, 5), 
                textcoords='offset points', fontsize=10)

# Centre de gravité
plt.scatter(centroid[0], centroid[1], c='orange', marker='D', s=100, 
           alpha=0.7, label='Centre de gravité')

plt.xlabel('Var1')
plt.ylabel('Var2')
plt.title('Classification K-moyennes - Partition en 2 classes')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"\n=== RÉSUMÉ ===")
print(f"Partition finale:")
print(f"  Classe 1: {[labels[i] for i in range(len(labels)) if cluster_labels[i] == 0]}")
print(f"  Classe 2: {[labels[i] for i in range(len(labels)) if cluster_labels[i] == 1]}")
print(f"Inertie intra-classe: {inertia:.1f}")
print(f"Pourcentage d'inertie expliquée: {percentage_explained:.1f}%")
```
